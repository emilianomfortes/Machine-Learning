## Exercise 2.1, Part 2: Sigmoid function

Before you start with the actual cost function, recall that the logistic regression hypothesis is defined as:

<center> $h_{\theta}(x)=g\left(\theta^{T} x\right)$
</center>
    
where g is the sigmoid function, which is defined by
    
<center> $g(z)=\frac{1}{1+e^{-z}}$
</center>
    
Your first step is to implement this function in **sigmoid** so it can be
called by the rest of your program. Evaluating sigmoid(0)
should give you exactly 0.5. Your code should also work with vectors and
matrices. For a matrix, your function should perform the sigmoid
function on every element.
    


```python
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))

#Element-wise version of the function for arrays
vsigmoid = np.vectorize(sigmoid) 

#Let's visualize it
I = np.linspace(-10,10,1000)
plt.figure()
plt.plot(I,sigmoid(I))
plt.xlabel('x'), plt.ylabel('sigmoid(x)')
```

## Exercise 2.1, Part 3: Cost function and gradient

Now you will implement the cost function and gradient for logistic regression.
Complete the code in *costFunction* to return the cost and gradient.

Recall that the cost function in logistic regression is


<center> $J(\theta)=\frac{1}{m} \sum_{i=1}^{m}\left[-y^{(i)} \log \left(h_{\theta}\left(x^{(i)}\right)\right)-\left(1-y^{(i)}\right) \log \left(1-h_{\theta}\left(x^{(i)}\right)\right)\right],$ <center>
    
    
and the gradient of the cost is a vector of the same length $\theta$ where the $j^{th}$ element (for $j=0,1,\dots,n$) is defined as follows:
    
    
<center> $\frac{\partial J(\theta)}{\partial \theta_{j}}=\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}$ <center>
    
    
Note that while this gradient looks identical to the linear regression gradient, the formula is actually different because linear and logistic regression have different definitions of $h_{\theta}(x)$

```python
def costFunction(Theta, X, Y):
    m = len(Y)
    
    #First we add a column of ones to X to fit the bias unit
    X = np.append(np.ones((X.shape[0],1)),X,axis=1)
    
    #Cost function
    J = (1/m) * (-Y.dot(np.log(vsigmoid(Theta.dot(X.T)))) - (1 - Y).dot(np.log(1 - vsigmoid(Theta.dot(X.T)))))
    #cost_history.append(J)
    
    #Gradient
    grad = (1/m) * (vsigmoid(Theta.dot(X.T)) - Y).dot(X)
    
    return J, grad
```

Once you are done, call your **costFunction** using the initial
parameters of $\theta$ equal to the null vector. You should see that the cost is about 0.693.

```python
X = data[['Exam 1 score','Exam 2 score']].to_numpy()
#X = np.append(np.ones((X.shape[0],1)),X,axis=1) # add a column of ones for theta0

m, n = X.shape #Number of examples, number of features

Y = data['Pass'].astype(int).to_numpy()
Y = Y.reshape(len(Y))

#Set initial theta for testing to the null vector
initial_theta = np.zeros((n+1))

cost, grad = costFunction(initial_theta,X,Y)
print(cost)
print(grad)
```

## Exercise 2.2, Part 2: Feature mapping

One way to fit the data better is to create more features from each data
point. You will code the function mapFeature, which will map the features into
all polynomial terms of $x_1$ and $x_2$ up to the sixth power.

<center>
    mapFeature $(x)=\left[\begin{array}{c}1 \\ x_{1} \\ x_{2} \\ x_{1}^{2} \\ x_{1} x_{2} \\ x_{2}^{2} \\ x_{1}^{3} \\ \vdots \\ x_{1} x_{2}^{5} \\ x_{2}^{6}\end{array}\right]$
<center>
    

As a result of this mapping, our vector of two features (the scores on
two QA tests) has been transformed into a 28-dimensional vector. A logistic
regression classifier trained on this higher-dimension feature vector will have
a more complex decision boundary and will appear nonlinear when drawn in
our 2-dimensional plot.
While the feature mapping allows us to build a more expressive classifier,
it also more susceptible to overfitting. In the next parts of the exercise, you
will implement regularized logistic regression to fit the data and also see for
yourself how regularization can help combat the overfitting problem.

*Remark 1*: We could define the feature mapping function ourselves, which is not hard at all, but scikit-learn already has an implemented version of the algorithm which we will make use in order to save time. The documentation of sklearn.preprocessing.PolynomialFeatures can be found in https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html

*Remark 2*: Because of the feature mapping, we do not need to add a bias unit feature, since it is obtained automatically with the polynomial features.

```python
#We could define the feature mapping, but sklearn has already a comfortable implementation which we make use of
from sklearn.preprocessing import PolynomialFeatures

def mapFeature_sklearn(X,degree):
    poly = PolynomialFeatures(degree, interaction_only=False, include_bias=True, order='C')
    return poly.fit_transform(X)
            
def mapFeatures(X, deg):
    X1 = X[:,0]
    X2 = X[:,1]
    if X1.ndim > 0:
        out = [np.ones(X1.shape[0])]
        for i in range(1, deg + 1):
            for j in range(i + 1):
                out.append((X1 ** (i - j)) * (X2 ** j))
        return np.stack(out, axis=1)
    else:
        out = [np.ones(1)]
        for i in range(1, deg + 1):
            for j in range(i + 1):
                out.append((X1 ** (i - j)) * (X2 ** j))
        return np.array(out)

#Data in pandas dataframe to numpy arrays
X = data[['Test microchip 1', 'Test microchip 2']].to_numpy()
Y = data['Passed'].to_numpy()

#Degree of polynomial feature mapping
degree = 6

#Map with own function
Xmap = mapFeatures(X,degree)
print('Resulting shape of the training examples with polynomical features using my own function')
print(Xmap.shape)

#Map with sklearn PolynomialFeatures
Xsklearn = mapFeature_sklearn(X,degree)
print('Resulting shape of the training examples with polynomical features using sklearn PolynomialFeatures function')
print(Xsklearn.shape)

#And we can compare both cases to see they are equal (numerically speaking)
print('Checking that both functions yield the same result')
print(np.sum(Xmap[3,:]-Xsklearn[3,:]))

m, n = Xmap.shape #number of training examples and features
```

## Exercise 2.2, Part 3: Regularized cost and gradient functions

Now you will implement code to compute the cost function and gradient for
regularized logistic regression. Complete the code in costFunctionReg.m to
return the cost and gradient.
Recall that the regularized cost function in logistic regression is 

<center>
$J(\theta)=\frac{1}{m} \sum_{i=1}^{m}\left[-y^{(i)} \log \left(h_{\theta}\left(x^{(i)}\right)\right)-\left(1-y^{(i)}\right) \log \left(1-h_{\theta}\left(x^{(i)}\right)\right)\right]+\frac{\lambda}{2 m} \sum_{j=1}^{n} \theta_{j}^{2},$
<center>


and the regularized gradient is

    
<center>
    $\frac{\partial J(\theta)}{\partial \theta_{0}}=\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)} \quad$ for $j=0$,
<center>
    
<center>
    $\frac{\partial J(\theta)}{\partial \theta_{j}}=\left(\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}\right)+\frac{\lambda}{m} \theta_{j} \quad$ for $j \geq 1.$
<center>
    
Once you are done, call your costFunctionReg function
using the initial value of $\theta$ (initialized to all zeros) and $\lambda = 1$. You should see that the
cost is about 0.693.

```python
def costFunctionReg(Theta, X, Y,Lambda):
    m = len(Y)
    
    #Cost function
    J = (1/m) * (-Y.dot(np.log(vsigmoid(Theta.dot(X.T)))) - (1 - Y).dot(np.log(1 - vsigmoid(Theta.dot(X.T)))))\
        + .5*(Lambda/m)*np.sum(Theta[1:]**2)
    
    #Gradient
    grad = (1/m) * (vsigmoid(Theta.dot(X.T)) - Y).dot(X) + (Lambda/m)*np.append(0,Theta[1:])
    
    return J, grad


#Cost at theta = null vector and lambda = 1
initial_theta = np.zeros(n)
cost, grad = costFunctionReg(initial_theta,Xmap,Y,1)
print('The cost function at (theta = null vector, regularization parameter = 1) is '\
      +str(cost))
```

## Exercise 2.2, Part 4: Learning parameters using fminunc
Similar to the previous parts, you will use fminunc to learn the optimal
parameters $\theta$. If you have completed the cost and gradient for regularized
logistic regression (costFunctionReg) correctly, you should be able to step
through the next part to learn the parameters $\theta$ using fminunc.

```python
options= {'maxiter': 400}
Lambda = 1

Theta_min = minimize(costFunctionReg,
                        initial_theta,
                        (Xmap, Y, Lambda),
                        jac=True,
                        method='Newton-CG',
                        options=options)

print('The value that minimizes the parameters is ')
print(Theta_min.x)

print('The associated cost value is')
print(costFunctionReg(Theta_min.x,Xmap,Y,0)[0])
```

## Exercise 2.2, Part 5: Plotting the decision boundary

To help you visualize the model learned by this classifier, we have ask you to plot the (non-linear)
decision boundary that separates the positive and negative examples. The idea is to plot the non-linear decision boundary by computing the classifier’s predictions on an evenly spaced grid and then draw
a contour plot of where the predictions change from y = 0 to y = 1.


```python
def decisionBoundary(x_1,x_2,theta,degree):
    xmap = np.append(x_1.reshape(1,1),x_2.reshape(1,1),axis=1)
    X = mapFeatures(xmap,degree)
    return np.sum(theta.dot(X.T))

def decisionBoundaryPlot(x1,x2,theta,degree):
    Z = np.zeros((100,100),dtype=float)
    for i in range(100):
        for j in range(100):
            Z[i,j] = decisionBoundary(x1[i],x2[j],theta,degree)
    return Z

x1 = np.linspace(-1.5,1.5,100)
x2 = np.linspace(-1.5,1.5,100)
X1, X2 = np.meshgrid(x1, x2)

plt.figure(figsize=(6,18))

# PLOT LEFT
plt.subplot(3,1,1)
Lambda = 0
plt.title(r"$\lambda = $"+str(Lambda))

Theta_min = minimize(costFunctionReg,initial_theta,(Xmap, Y, Lambda),jac=True,method='Newton-CG',options=options)
Z = decisionBoundaryPlot(x1,x2,Theta_min.x,6)

sns.scatterplot(x=data['Test microchip 1'],y=data['Test microchip 2'], style=data['Passed'],hue=data['Passed'],s=100)
plt.contour(X1,X2,Z,colors='magenta', levels=[0.5])
plt.xlim(-1,1), plt.ylim(-1,1.2)

#PLOT MIDDLE
plt.subplot(3,1,2)
Lambda = 1
plt.title(r"$\lambda = $"+str(Lambda))

Theta_min = minimize(costFunctionReg,initial_theta,(Xmap, Y, Lambda),jac=True,method='Newton-CG',options=options)
Z = decisionBoundaryPlot(x1,x2,Theta_min.x,6)

sns.scatterplot(x=data['Test microchip 1'],y=data['Test microchip 2'], style=data['Passed'],hue=data['Passed'],s=100)
plt.contour(X1,X2,Z,colors='magenta', levels=[0.5])
plt.xlim(-1,1), plt.ylim(-1,1.2)

#PLOT RIGHT
plt.subplot(3,1,3)
Lambda = 5
plt.title(r"$\lambda = $"+str(Lambda))


Theta_min = minimize(costFunctionReg,initial_theta,(Xmap, Y, Lambda),jac=True,method='Newton-CG',options=options)
Z = decisionBoundaryPlot(x1,x2,Theta_min.x,6)

sns.scatterplot(x=data['Test microchip 1'],y=data['Test microchip 2'], style=data['Passed'],hue=data['Passed'],s=100)
plt.contour(X1,X2,Z,colors='magenta', levels=[0.5])
plt.xlim(-1,1), plt.ylim(-1,1.2)
```
