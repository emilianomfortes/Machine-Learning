## Exercise 3.1, Part 2: Vectorizing Logistic Regression

You will be using multiple one-vs-all logistic regression models to build a
multi-class classifier. Since there are 10 classes, you will need to train 10
separate logistic regression classifiers. To make this training efficient, it is
important to ensure that your code is well vectorized. In this section, you
will implement a vectorized version of logistic regression that does not employ
any for loops. You can use your code in the last exercise as a starting point
for this exercise.

Recall that in logistic regression, the cost function is

<center>
    $J(\theta)=\frac{1}{m} \sum_{i=1}^{m}\left[-y^{(i)} \log \left(h_{\theta}\left(x^{(i)}\right)\right)-\left(1-y^{(i)}\right) \log \left(1-h_{\theta}\left(x^{(i)}\right)\right)\right] + \lambda \sum_{j=1}^n \theta^2_j.$
<center>
    
To compute each element in the summation, and for every example i, we have to compute
    
<center>
    $h_{\theta}(x^{(i)}) = \frac{1}{1+e^{-\theta^{T}x^{(i)}}}$ 
<center>
    
It turns out that we can compute this quickly for all our
examples by using matrix multiplication. Let us define $X$ and $\theta$ as    
    
<center>
    $X=\left[\begin{array}{c}-\left(x^{(1)}\right)^{T}- \\ -\left(x^{(2)}\right)^{T}- \\ \vdots \\ -\left(x^{(m)}\right)^{T}-\end{array}\right] \qquad \text{and} \qquad \theta=\left[\begin{array}{c}\theta_{0} \\ \theta_{1} \\ \vdots \\ \theta_{n}\end{array}\right]$ 
<center>
    
Then, by computing the matrix product $X\theta$, we have
    
<center>
    $X \theta=\left[\begin{array}{c}-\left(x^{(1)}\right)^{T} \theta- \\ -\left(x^{(2)}\right)^{T} \theta- \\ \vdots \\ -\left(x^{(m)}\right)^{T} \theta-\end{array}\right]=\left[\begin{array}{c}-\theta^{T}\left(x^{(1)}\right)- \\ -\theta^{T}\left(x^{(2)}\right)- \\ \vdots \\ -\theta^{T}\left(x^{(m)}\right)-\end{array}\right]$
<center>

Your job is to write the unregularized cost function in the file **lrCostFunction**
Your implementation should use the strategy we presented above to calculate $\theta^T$ x^{(i)}$. You should also use a vectorized approach for the rest of the
cost function.

In order to vectorize the gradient, we have to build up a code such that

<center>
$\left[\begin{array}{c}\frac{\partial J}{\partial \theta_{0}} \\ \frac{\partial J}{\partial \theta_{1}} \\ \frac{\partial J}{\partial \theta_{2}} \\ \vdots \\ \frac{\partial J}{\partial \theta_{n}}\end{array}\right]=\frac{1}{m}\left[\begin{array}{c}\sum_{i=1}^{m}\left(\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{0}^{(i)}\right) \\ \sum_{i=1}^{m}\left(\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{1}^{(i)}\right) \\ \sum_{i=1}^{m}\left(\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{2}^{(i)}\right) \\ \vdots \\ \sum_{i=1}^{m}\left(\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{n}^{(i)}\right)\end{array}\right] + \frac{\lambda}{m}\left[\begin{array}{c}0 \\ \theta_1 \\  \theta_2 \\ \vdots \\ \theta_n\end{array}\right]$
<center>

```python
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))

#Element-wise version of the function for arrays
vsigmoid = np.vectorize(sigmoid) 


def lrCostFunction(Theta, X, Y, Lambda):
    m = len(Y)
    
    #First we add a column of ones to X to fit the bias unit
    X = np.append(np.ones((X.shape[0],1)),X,axis=1)
    
    #Cost function
    J = (1/m) * (-Y.dot(np.log(vsigmoid(Theta.dot(X.T)))) - (1 - Y).dot(np.log(1 - vsigmoid(Theta.dot(X.T)))))\
        + .5*(Lambda/m)*np.sum(Theta[1:]**2)
    
    #Gradient
    grad = (1/m) * (vsigmoid(Theta.dot(X.T)) - Y).dot(X) + (Lambda/m)*np.append(0,Theta[1:])
 
    return J, grad

init_theta = np.zeros(n+1) #+1 for the bias unit
print(lrCostFunction(init_theta,X,Y,0)[0])
print(len(lrCostFunction(init_theta,X,Y,0)[1]))
```

## Exercise 3.1, Part 3: One-vs-all Classification

In this part of the exercise, you will implement one-vs-all classification by
training multiple regularized logistic regression classifiers, one for each of
the $K$ classes in our dataset (Figure 1). In the handwritten digits dataset,
$K = 10$, but your code should work for any value of $K$.

You should now complete the code **oneVsAll** to train one classifier for
each class. In particular, your code should return all the classifier parameters
in a matrix $\Theta\in\mathbb{R}^{K\times(n+1)}$, where each row of $\Theta$ corresponds to the learned
logistic regression parameters for one class. You can do this with a for-loop
from 1 to $K$, training each classifier independently.
Note that the y argument to this function is a vector of labels from 1 to
10.

Furthermore, you will be using fmincg for this exercise (instead of fminunc).
fmincg works similarly to fminunc, but is more more efficient for dealing with
a large number of parameters.

```python
def oneVsAll(Theta0, X, Y, Lambda):
    K, counts = np.unique(Y,return_counts=True)
    m, n = X.shape
    Thetas_min = np.zeros((len(K),n+1))
    for k in K:
        Y_k = np.array(list(map(lambda x: 1 if x==k else 0, Y)))
        Thetas_min[k-1,:] = minimize(lrCostFunction,Theta0[k-1,:],(X, Y_k, Lambda), jac=True, 
                                     method='L-BFGS-B',options={'maxiter':400}).x
        print('Done with iteration '+str(k)) #'Newton-CG' DOES NOT WORK
    return Thetas_min
        
Lambda = 1
Theta0 = np.zeros((len(K),n+1))
#print(Theta0[k-1,:].shape)
#print(init_theta.shape)
Thetas_min = oneVsAll(Theta0, X, Y, Lambda)
```

## Exercise 3.1, Part 4: One-vs-all Prediction

After training your one-vs-all classifier, you can now use it to predict the
digit contained in a given image. For each input, you should compute the
probability that it belongs to each class using the trained logistic regression
classifiers.

Your one-vs-all predict function will pick the class for which the
corresponding logistic regression classifier outputs the highest probability and
return the class label (1, 2,..., or K) as the prediction for the input example.
You should now complete the code in **predictOneVsAll** to use the
one-vs-all classifier to make predictions.

Once you are done, call your **predictOneVsAll** function using
the learned value of $\Theta$. You should see that the training set accuracy is about
94.9% (i.e., it classifies 94.9% of the examples in the training set correctly).

```python
def predictOneVsAll(Thetas, X):
    X = np.append(np.ones((X.shape[0],1)),X,axis=1)
    sigmoides = vsigmoid(Thetas.dot(X.T))
    return sigmoides.argmax(axis=0) + 1

def accuracyOneVsAll(Thetas, X, Y):
    return np.sum(predictOneVsAll(Thetas,X) == Y)/len(Y)

print('accuracy obtained = '+str(accuracyOneVsAll(Thetas_min,X,Y))+' / 1')
print('acuracy expected = 0.949')
```

## Exercise 3.2, Part 1: Feedforward Propagation and Prediction

Now you will implement feedforward propagation for the neural network. You
will need to complete the code in **predict** to return the neural networkâ€™s
prediction. 

You should implement the feedforward computation that computes $h_{\theta}(x^{(i)})$
for every example i and returns the associated predictions. Similar to the
one-vs-all classification strategy, the prediction from the neural network will
be the label that has the largest output $h_{\theta}(x)_k$.

Once you are done, call your **predict** function using the
loaded set of parameters for Theta1 and Theta2. You should see that the accuracy is about 97.5%.

```python
def NN_feedforward(Theta1,Theta2,X):
    a1 = np.append(np.ones((X.shape[0],1)),X,axis=1)
    z2 = Theta1.dot(a1.T).T
    a2 = vsigmoid(z2)
    a2 = np.append(np.ones((a2.shape[0],1)),a2,axis=1)
    z3 = Theta2.dot(a2.T).T
    a3 = vsigmoid(z3)
    return a3

def NN_predictOneVsAll(a3):
    return a3.argmax(axis=1) + 1

def NN_accuracyOneVsAll(a3,Y):
    return np.sum(NN_predictOneVsAll(a3) == Y)/len(Y)

a3 = NN_feedforward(Theta1,Theta2,X)

print('accuracy obtained = '+str(NN_accuracyOneVsAll(a3,Y))+' / 1')
print('acuracy expected = 0.975')
```
