## Exercise 5.1, Part 2: Regularized linear regression cost function

Recall that regularized linear regression has the following cost function:

<center>
    $J(\theta)=\frac{1}{2 m}\left(\sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right)^{2}\right)+\frac{\lambda}{2 m}\left(\sum_{j=1}^{n} \theta_{j}^{2}\right),$
<center>
<br>
where $\lambda$ is a regularization parameter which controls the degree of regularization (thus, help preventing overfitting). The regularization term puts a
penalty on the overal cost $J$. As the magnitudes of the model parameters $\theta_j$
increase, the penalty increases as well. Note that you should not regularize
the $\theta_0$ term. 
   
Correspondingly, the partial derivative of regularized linear regression’s cost for θj is defined as    
    
<center>
    $\begin{aligned} \frac{\partial J(\theta)}{\partial \theta_{0}} &=\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)} & & \text { for } j=0 \\ \frac{\partial J(\theta)}{\partial \theta_{j}} &=\left(\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}\right)+\frac{\lambda}{m} \theta_{j} & \text { for } j \geq 1 \end{aligned}$
<center>
<br>
    
You should now complete the code in the function **linearRegCostFunction**. When you are finished, run your cost function
using theta initialized at [1; 1]. You should expect to see an output of 303.993 and a gradient of [-15.30; 598.250].

```python
def linearRegCostFunction(Theta,X,y,Lambda):
    m = len(y)
    J = (.5/m)* np.sum(np.power(Theta.dot(X.T)-y,2)) + .5*(Lambda/m)*np.sum(Theta[1:]**2)
    grad = (1/m)*X.T.dot((Theta.dot(X.T)-y)) + (Lambda/m)*np.append(0,Theta[1:])
    return J, grad

initial_theta = np.ones((2))
initial_lambda = 1

print('You should expect to see an output of 303.993 and a gradient of [-15.30; 598.250].')
print('Result:')
print(linearRegCostFunction(initial_theta,X,y,initial_lambda))
#print(linearRegCostFunction(initial_theta,X,y,0)[1].ravel())
```

## Exercise 5.1, Part 3: Fitting linear regression

Once your cost function and gradient are working correctly, the next part of
the code should compute the optimal values of $\theta$. This training function uses fmincg to optimize the cost function.

In this part, we set regularization parameter $\lambda$ to zero. Because our
current implementation of linear regression is trying to fit a 2-dimensional $\theta$,
regularization will not be incredibly helpful for a $\theta$ of such low dimension. In
the later parts of the exercise, you will be using polynomial regression with
regularization.


Finally, thescript should also plot the best fit line, resulting in
an image similar to the original data visualization Figure. The best fit line tells us that the model is

```python
Theta_min = minimize(linearRegCostFunction,
                        initial_theta,
                        (X, y, 0),
                        jac=True,
                        method='TNC',
                        options={'maxiter': 400})

#Visualization

plt.figure(figsize=(8,6))

plt.scatter(X[:,1],y,color="deeppink")

x = np.linspace(min(data['X']),max(data['X']),100)
plt.plot(x,Theta_min.x[0] + Theta_min.x[1]*x)

plt.grid(True)
plt.xlabel("Change in water level (x)"), plt.ylabel("Water flowing out of the dam (y)"), plt.title("Data visualization")
plt.show()
```

## Exercise 5.1, Part 4: Bias-variance

An important concept in machine learning is the bias-variance tradeoff. Models with high bias are not complex enough for the data and tend to underfit,
while models with high variance overfit to the training data.

In this part of the exercise, you will plot training and test errors on a
learning curve to diagnose bias-variance problems.

You will now implement code to generate the learning curves that will be
useful in debugging learning algorithms. Recall that a learning curve plots
training and cross validation error as a function of training set size. Your
job is to fill in **learningCurve** so that it returns a vector of errors for the
training set and cross validation set.

To plot the learning curve, we need a training and cross validation set
error for different training set sizes. To obtain different training set sizes,
you should use different subsets of the original training set X. Specifically, for
a training set size of i, you should use the first i examples (i.e., X(1:i,:)
and y(1:i))

After learning the θ parameters, you should compute the error on the training and cross validation sets. Recall that the training error for a dataset is
defined as

<center>
    $J_{\text {train }}(\theta)=\frac{1}{2 m}\left[\sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right)^{2}\right]$
</center>
<br>
In particular, note that the training error does not include the regularization term. One way to compute the training error is to use your existing
cost function and set λ to 0 only when using it to compute the training error
and cross validation error. When you are computing the training set error,
make sure you compute it on the training subset (i.e., X(1:n,:) and y(1:n))
(instead of the entire training set). However, for the cross validation error,
you should compute it over the entire cross validation set. You should store
the computed errors in the vectors **error_train** and **error_val**.

```python
def learningCurve(Theta,X,y,Xval,yval,Lambda):
    m, n = X.shape
    error_train = np.zeros(m)
    error_val = np.zeros(m)
    for i in range(1,m+1):
        Theta_min = minimize(linearRegCostFunction,
                            initial_theta,
                            (X[:i,:], y[:i], 0),
                            jac=True,
                            method='TNC',
                            options={'maxiter': 400})
        error_train[i-1] = linearRegCostFunction(Theta_min.x,X[:i,:],y[:i],0)[0]
        error_val[i-1] = linearRegCostFunction(Theta_min.x,Xval,yval,0)[0]
    return error_train, error_val

Xval = data['Xval'] #shape=(len(Xval),1)
#We add a column of ones to fit theta_0 (the bias unit)
Xval = np.append(np.ones((Xval.shape[0],1)),Xval,axis=1) #shape=(len(Xval),2)

yval = data['yval'].ravel() #shape=(len(Xval),1)

error_train, error_val = learningCurve(initial_theta,X,y,Xval,yval,0)

iterations = np.arange(1,m+1,1)

plt.figure(figsize=(8,6))
plt.plot(iterations,error_train,label='Train error')
plt.plot(iterations,error_val,label='Cross validation error')
plt.legend(loc="best")
```

# Exercise 5.2: Polynomial regression

The problem with our linear model was that it was too simple for the data
and resulted in underfitting (high bias). In this part of the exercise, you will
address this problem by adding more features.
For use polynomial regression, our hypothesis has the form:

<center>
$\begin{aligned} h_{\theta}(x) &=\theta_{0}+\theta_{1} *(\text { waterLevel })+\theta_{2} *(\text { waterLevel })^{2}+\cdots+\theta_{p} *(\text { waterLevel })^{p} \\ &=\theta_{0}+\theta_{1} x_{1}+\theta_{2} x_{2}+\ldots+\theta_{p} x_{p} \end{aligned}$ </center><br>

Notice that by defining $x1 = (waterLevel)$; $x2 = (waterLevel)^2$, $\dots$, $xp = (waterLevel)^p$, we obtain a linear regression model where the features are the
various powers of the original value $(waterLevel)$.

Now, you will add more features using the higher powers of the existing
feature x in the dataset. Your task in this part is to complete the code in
**polyFeatures** so that the function maps the original training set $X$ of size
$m\times1$ into its higher powers. Specifically, when a training set $X$ of size $m \times 1$
is passed into the function, the function should return a $m\times p$ matrix $X$ poly,
where column 1 holds the original values of $X$, column 2 holds the values of
$X$.^2, column 3 holds the values of $X$.^3, and so on. Note that you don’t
have to account for the zero-eth power in this function.
Now you have a function that will map features to a higher dimension,
and apply it to the training set, the test set, and the
cross validation set (which you haven’t used yet).

Keep in mind that even though we have polynomial terms in our feature
vector, we are still solving a linear regression optimization problem. The
polynomial terms have simply turned into features that we can use for linear
regression. We are using the same cost function and gradient that you wrote
for the earlier part of this exercise.
For this part of the exercise, you will be using a polynomial of degree 8.
It turns out that if we run the training directly on the projected data, will
not work well as the features would be badly scaled (e.g., an example with
x = 40 will now have a feature x8 = 408 = 6:5 × 1012). Therefore, you will
need to use feature normalization.

Before learning the parameters $\theta$ for the polynomial regression, first we have to define and call a **featureNormalize** function which normalizes the features of the training set, as well as storing the $\mu$ and $\sigma$ parameters separately.

```python
X = data['X'].ravel() #shape=(12,1)

def polyFeatures(X,degree):
    Xd = np.zeros((len(X),degree+1))
    for i in range(degree+1):
        Xd[:,i] = np.power(X[:],i)
    return Xd

def featureNormalize(X):
    m, n = X.shape
    mu = np.zeros(n-1)
    sigma = np.zeros(n-1)
    for i in range(1,n):
        mu[i-1] = np.mean(X[:,i])
        sigma[i-1] = np.std(X[:,i])
        X[:,i] = (X[:,i]-mu[i-1])/sigma[i-1]
    return X, mu, sigma

def gridNormalize(X,mu,sigma):
    m, n = X.shape
    for i in range(1,n):
        X[:,i] = (X[:,i]-mu[i-1])/sigma[i-1]
    return X

degree = 8
Xpoly = polyFeatures(X,degree)
Xpoly_norm, mu, sigma = featureNormalize(Xpoly)
#print(Xpoly_norm[:,6:9])
```

## Exercise 5.2, Part 1: Learning Polynomial Regression

After normalizing the training examples, you should learn the parameters with the minimization method and plot the results in a figure.

```python
initial_theta_poly = np.zeros(degree+1)

Theta_min_poly = minimize(linearRegCostFunction,
                        initial_theta_poly,
                        (Xpoly_norm, y, 0),
                        jac=True,
                        method='TNC',
                        options={'maxiter':200})
print(Theta_min_poly.x)
#Visualization

plt.figure(figsize=(8,6))

plt.scatter(X,y,color="deeppink")

plot_range=1000
x = np.linspace(-65,65,plot_range).ravel()
xpoly = polyFeatures(x,degree)
xpoly_norm = gridNormalize(xpoly, mu, sigma)

#xpoly_norm = 
#xpoly_norm = featureNormalize(xpoly)[0]
plt.plot(x,xpoly_norm.dot(Theta_min_poly.x))

plt.grid(True)
plt.xlabel("Change in water level (x-mu)/sigma"), plt.ylabel("Water flowing out of the dam (y)"), plt.title("Data visualization")
plt.show()
```
